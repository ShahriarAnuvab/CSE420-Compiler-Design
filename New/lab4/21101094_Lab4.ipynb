```python
class ParseStack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def peek(self):
        if not self.is_empty():
            return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

parse_stack = ParseStack()

terminals = ["id", "*", "+", "(", ")", "$"]
production_rules = {
    "id": {"E": "TE'", "T": "FT'", "F": "id"},
    "+": {"E'": "+TE'", "T'": "epsilon","T": "synch", "F": "synch"},
    "*": {"T'": "*FT'", "F": "synch"},
    "(": {"E": "TE'", "T": "FT'", "F": "(E)"},
    ")": {"E'": "epsilon", "T'": "epsilon","E": "synch","T": "synch","F": "synch"},
    "$": {"E'": "epsilon", "T'": "epsilon","E": "synch","T": "synch","F": "synch"}
}

input_string = str(input("Enter your string ")).lower()
input_string += "$"
input_tokens = []

while input_string:
    for term in terminals:
        if input_string.startswith(term):
            input_tokens.append(term)
            input_string = input_string[len(term):]
            break

parse_stack.push("$")
parse_stack.push("E")
input_pointer = 0

matched_tokens = []
print(f"{' Matched':<20}{'Stack':<20}{'Input':<20}{'Action':<20}")
print("-------------------------------------------------------------------")
print(f"{'':<20}{''.join(reversed(parse_stack.items)):<20}{''.join(input_tokens[input_pointer:]):<20}{'':<20}")

while True:
    x = parse_stack.peek()
    a = input_tokens[input_pointer]

    if x in terminals or x == "$":
        if x == a:
            if x == "$":
                print()
                print("Parsing Successful")
                break
            parse_stack.pop()
            matched_tokens.append(a)
            print(f"{''.join(matched_tokens):<20}{''.join(reversed(parse_stack.items)):<20}{''.join(input_tokens[input_pointer+1:]):<20}{'match ' + a:<20}")
            input_pointer += 1
        else:
            parse_stack.pop()
    else:
        if production_rules[a][x] != "":
            rule = production_rules[a][x]
            q = parse_stack.pop()

            if rule in terminals:
                parse_stack.push(rule)
                print(f"{''.join(matched_tokens):<20}{''.join(reversed(parse_stack.items)):<20}{''.join(input_tokens[input_pointer:]):<20}{q + ' ==> ' + rule:<20}")
            elif rule == "epsilon":
                print(f"{''.join(matched_tokens):<20}{''.join(reversed(parse_stack.items)):<20}{''.join(input_tokens[input_pointer:]):<20}{q + ' ==> ' + rule:<20}")
                pass
            elif rule == "synch":
                print(f"{''.join(matched_tokens):<20}{''.join(reversed(parse_stack.items)):<20}{''.join(input_tokens[input_pointer:]):<20}{q + ' ==> ' + rule:<20}")
                pass
            else:
                i = len(rule) - 1
                while i >= 0:
                    if rule[i] == "'":
                        combined_symbol = rule[i - 1] + rule[i]
                        parse_stack.push(combined_symbol)
                        i -= 2
                    else:
                        parse_stack.push(rule[i])
                        i -= 1
                print(f"{''.join(matched_tokens):<20}{''.join(reversed(parse_stack.items)):<20}{''.join(input_tokens[input_pointer:]):<20}{q + ' ==> ' + rule:<20}")
        else:
            input_pointer += 1
```
